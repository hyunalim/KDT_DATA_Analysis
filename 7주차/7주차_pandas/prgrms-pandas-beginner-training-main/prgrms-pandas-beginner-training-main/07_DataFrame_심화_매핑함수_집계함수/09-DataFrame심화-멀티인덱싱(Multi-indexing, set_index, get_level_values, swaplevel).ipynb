## 멀티인덱스? 멀티인덱싱?
- https://pandas.pydata.org/docs/reference/api/pandas.MultiIndex.html#pandas.MultiIndex
- 하나 이상의 인덱스 레벨을 가지는 인덱스 구조.
- 데이터프레임이나 시리즈의 다차원적인 인덱싱을 지원하기 위해 사용.
- 멀티인덱스로 인해 데이터를 계층적(hierarchical)으로 조직화.
- 다양한 차원(Multi-level)에서 데이터에 접근 가능.
- 멀티인덱스 생성은 데이터프레임 또는 시리즈를 생성할 때 인덱스를 array같은 형태로 지정한다.

## 멀티인덱스를 활용해 조회하기

```python
# 초기화
customers = pd.read_csv('marketing_campaign.csv', sep='\t', index_col=['Marital_Status', 'ID'])
customers
```

```python
# Select row - index level 0
customers.loc['Single']
```

```python
# Select row - columns
customers.loc['Single','Income']
```

```python
customers.loc[('Single',5524),'Income']
```

```python
customers.loc[('Single',5524)]
```

```python
customers.loc[('Single',5524),('Income','Kidhome')]
```

## 슬라이싱(:)을 이용한 인덱스 여러개 값 조회

```python
# 'Single', 'Alone', 'Absurd', 'YOLO' 만 조회
customers.loc[['Single', 'Alone', 'Absurd', 'YOLO']]
customers.loc[['Single', 'Alone', 'Absurd', 'YOLO'],:]
```

```python
# 2개의 인덱스를 사용할 땐
# .loc[([lv1_labels], [lv2_labels])]
# 'Single', 'Alone', 'Absurd', 'YOLO'
# 5524, 2114, 92, 4369, 11133
customers.loc[(['Single','Alone','Absurd','YOLO'],[5524,2114,92,4369,11133]),]
```

## 특정 level 생략 - 행 조회

```python
# 특정 row 조회
customers.loc[:,2114,:]
```

```python
# 특정 여러 row 조회
customers.loc[:,[2114,92],:]
```

```python
# 뒤에 :를 생략하면 에러
customers.loc[:,[2114,92],]
```

## 특정 level 생략 - 행과 열 조회

```python
# 하나씩 해도 에러
customers.loc[:,2114,'Income']
```

```python
# 특정 row - col 조회
customers.loc[(:,2114),'Income']
```

```python
# 위에서 본 특정 여러 row 조회에서 조금 변경해보면? Try 1
customers.loc[(:,[2114,93]),('Income','Kidhome')]
```

```python
# 위에서 본 특정 여러 row 조회에서 조금 변경해보면? Try 2
```

```python
# 해결 방법 1
customers.loc[(slice(None),2114),'Income']
```

```python
customers.loc[(slice(None),2114),['Income','Kidhome']]
```

```python
# 해결 방법 2
customers.loc[pd.IndexSlice[:,2114],['Income','Kidhome']]
```

```python
customers.loc[pd.IndexSlice[:,2114],'Income']
```
