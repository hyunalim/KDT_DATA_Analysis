## 이름(name) 변경

```python
# 참조
import pandas as pd
```

```python
# 데이터프레임 생성
cols = ['name', 'survived', 'pclass', 'fare', 'sex', 'age']
df = pd.read_excel('titanic3.xls', usecols=cols, index_col='name')
df
```

## 열(column) 이름 변경

```python
# 열 목록(columns index) 조회
df.columns
```

```python
# 특정 열 이름 접근
df.columns[0]
```

```python
# Try : 수정 = 접근 후 할당 
df.columns[0]='class'
```

```python
# 방법 1 : rename을 이용하기
df.rename(columns={'pclass':'class'}, inplace=True)
```

```python
df
```

```python
# 방법 2 : 같은 길이의 리스트를 준비해서 교체하기
df.columns
```

```python
# 길이가 5이니 5개 요소의 리스트 준비하기, 비용(fare->cost)로 변경
df.columns=['class','survived','sex','age','cost']
```

```python
df.columns
```

```python
df.head()
```

## rename() 좀 더 살펴보기 - 행과 열의 인덱스 이름 변경
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename.html
- https://pandas.pydata.org/docs/reference/api/pandas.Series.rename.html
- https://pandas.pydata.org/docs/reference/api/pandas.Index.rename.html

```python
# 1. 2개의 열을 변경하려면?
df.rename({'class':'pclass',
          'cost':'fare'}, axis=1)
```

```python
# axis인자를 빼고 한다면
df.rename(columns={'class':'pclass',
          'cost':'fare'})
```

```python
# 2. 인덱스에서 'Allen, Miss. Elisabeth Walton' 이름 변경, axis=0이 디폴트
df.rename({'Allen, Miss. Elisabeth Walton':'Allen'},axis=0)
```

```python
# axis인자를 빼고 한다면
df.rename(index={'Allen, Miss. Elisabeth Walton':'Allen'})
```

```python
# 3. 행렬 인덱스를 설정
df.index.name = 'passenger'
df.columns.name = 'information'
```
