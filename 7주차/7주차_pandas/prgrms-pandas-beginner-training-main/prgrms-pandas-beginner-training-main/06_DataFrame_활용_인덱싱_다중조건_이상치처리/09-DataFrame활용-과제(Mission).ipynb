## Pandas참조 후 titanic 데이터를 가져오기
- name열이 index로 설정

```python
import pandas as pd
```

```python
df= pd.read_excel('titanic3.xls', index_col='name')
df
```

## 데이터프레임의 columns만 조회

```python
df.columns
```

## .drop()을 이용해서 2개의 열을 삭제
- home.dest
- embarked
- sibsp
- 원본에 반영

```python
df.drop(columns=['home.dest','embarked','sibsp'], inplace=True)
```

## del을 이용한 열 삭제
- parch

```python
del df['parch']
```

## 사망자 명단 조회하기

```python
df[df['survived']==0]
```

## 시신 식별번호(body)가 없는 데이터 조회하기

```python
no_body_mask = df['body'].isnull()
df[no_body_mask]
```

살아있는 사람은 시신 식별번호가 없는 것은 당연

## 사망자이나 시신식별번호가 없는 명단 찾기

```python
not_survived_mask = df['survived']==0
df[not_survived_mask&no_body_mask]
```

사망자가 809명인데, 식별번호가 없는 명단이 688명이나 된다.

## 사망자이나 시신식별번호가 없는 명단으로 새로운 df 만들기
- `notfound_list`

```python
notfound_list = df[no_body_mask & not_survived_mask]
notfound_list
```

## notfound_list의 승객 등급별 개수 확인

```python
#모법답안
notfound_list['pclass'].value_counts()
```

```python
#내 답안
pclass1_mask = notfound_list['pclass']==1
pclass2_mask = notfound_list['pclass']==2
pclass3_mask = notfound_list['pclass']==3
```

```python
pclass1_mask.sum()
```

```python
pclass2_mask.sum()
```

```python
pclass3_mask.sum()
```

3등급(하류층)이 많으나, 이건 당연할 수 있다. 비율을 구해보자.

```python
(pclass3_mask.sum()/notfound_list['pclass'].sum() * 100).round(3)
```

## 전체 titanic 전체 승객 등급별 개수 확인

```python
#모범답안
df['pclass'].value_counts()
```

```python
#내답안
p1_mask = df['pclass']==1
p2_mask = df['pclass']==2
p3_mask = df['pclass']==3
```

```python
p1_mask.sum()
```

```python
p2_mask.sum()
```

```python
p3_mask.sum()
```

## 승객 등급별 실종 비율 확인

```python
#모범답안
notfound_list['pclass'].value_counts()/df['pclass'].value_counts()
```

```python
#내답안
(pclass1_mask.sum()/p1_mask.sum()*100).round(2)
```

```python
(pclass2_mask.sum()/p2_mask.sum()*100).round(2)
```

```python
(pclass3_mask.sum()/p3_mask.sum()*100).round(2)
```

## 1. 사망자의 하류층 티겟 가격 분포와 2. 생존자의 하류층 티켓 가격 분포를 확인해보자.

```python
#1. 사망자의 하류층 티겟 가격 분포와
p3_not_survived=df[(df['survived']==0)& (df['pclass']==3)]
p3_not_survived['fare'].describe()
```

```python
# 2. 생존자의 하류층 티켓 가격 분포
p3_survived=df[(df['survived']==1)& (df['pclass']==3)]
p3_survived['fare'].describe()
```

3등급에서 티켓가격의 차이가 생존유무와 관련성이 있어보이지 않는다.
